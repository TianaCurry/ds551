````python
aqi = aqi[["Year","Month","Day","Date","Country","Status","AQI Value"]]
aqi.head()
````

````python
aqi_avg = aqi.groupby(["Country","Year","Month"]).mean(numeric_only= True)
aqi_avg = aqi_avg.rename(columns={"AQI Value":"Avg AQI"}).reset_index()
aqi_avg
````
